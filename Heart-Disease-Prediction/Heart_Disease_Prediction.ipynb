```python
#Import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, RandomizedSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from scipy.stats import probplot
from feature_engine.variable_transformers import PowerTransformer
from sklearn.metrics import classification_report
```

```python
#Read the data
data = pd.read_csv('heart.csv')
```

```python
#Checking the head
data.head()
```

```python
#Shape of the data
data.shape
```

```python
#Checking the percentage of missing data
data.isnull().mean()*100
```

```python
#Lets visualise data uing pair plot
sns.pairplot(data)
```

```python
#Visualize the correlation among the features using Heat map
plt.figure(figsize=(20,10))
sns.heatmap(data.corr(), annot=True)
```

```python
#Seggregate the data to dependent and independent variable
X = data[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach',
       'exang', 'oldpeak', 'slope', 'ca', 'thal']]
y = data['target']
```

```python
#Split the data to training and testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
#Getting the shape
X_train.shape, X_test.shape
```

```python
#method to visualise the distrbution of the data
def diagonastic_plot(data, column):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.distplot(data[column])
    
    plt.subplot(1,2,2)
    probplot(x = data[column], plot=plt)
    plt.show()
```

```python
#Visualise the training data
for col in X_train.columns:
    diagonastic_plot(X_train,col)
```

```python
#Apply the Log-transformer for more  Gaussian Approximation
transformer = PowerTransformer(exp=1.5)
transformer.fit(X_train)
```

```python
#Transform the varibles
X_train = transformer.transform(X_train)
X_test = transformer.transform(X_test)
```

```python
#Visualise the training data after the transformation
for col in X_train.columns:
    diagonastic_plot(X_train,col)
```

```python
#Standardize the data
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
#Transform the data
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#Applying the Logistic Regression model
classifier = LogisticRegression(C=0.01, class_weight='balanced', dual=False,
                   fit_intercept=True, intercept_scaling=1, l1_ratio=None,
                   max_iter=120, multi_class='multinomial', n_jobs=None,
                   penalty='l2', random_state=None, solver='sag', tol=0.0001,
                   verbose=0, warm_start=True)
```

```python
#Fit the model
classifier.fit(X_train,y_train)
```

```python
#Predict the test set result
y_pred = classifier.predict(X_test)
```

```python
#Print the classification_report
print(classification_report(y_test, y_pred))
```

```python
#Make parameters for RandomizedearchCV
params = {
    'C':[0.01, 0.1, 1, 2, 3],
    'fit_intercept':[False, True],
    'solver':['newton-cg','lbfgs','sag','saga'],
    'max_iter' :[100,120,150,200,250],
    'multi_class':['auto','ovr','multinomial'],
    'warm_start' : [True, False]
}
```

```python
#Applying RandomizedearchCV
cross_val = RandomizedSearchCV(estimator=classifier,param_distributions=params, n_iter=10, cv=10,scoring='f1',
                              n_jobs=-1)
cross_val.fit(X_train,y_train)
```
